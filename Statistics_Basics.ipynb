### 1. **Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.**  
   - **Qualitative Data**: Non-numerical data describing categories or characteristics.  
     - Examples: Eye color, gender, blood type.  
     - Scales:  
       - **Nominal**: Categories with no specific order (e.g., gender: male, female).  
       - **Ordinal**: Categories with a meaningful order but no fixed interval (e.g., survey satisfaction levels: good, average, poor).  

   - **Quantitative Data**: Numerical data representing quantities.  
     - Examples: Age, height, income.  
     - Scales:  
       - **Interval**: Numerical data with no true zero, equal intervals (e.g., temperature in Celsius).  
       - **Ratio**: Numerical data with a true zero, equal intervals (e.g., weight, income).

### 2. **What are the measures of central tendency, and when should you use each?**  
   - **Mean**: The average of all data points. Use when data is symmetric and has no extreme outliers.  
     - Example: Average test scores.  
   - **Median**: The middle value when data is ordered. Use when data is skewed or has outliers.  
     - Example: Median income in a region.  
   - **Mode**: The most frequently occurring value. Use for categorical data or finding the most common value.  
     - Example: Favorite ice cream flavor in a survey.

---

### 3. **Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?**  
   - **Dispersion**: Refers to how spread out data points are around the central value.  
   - **Variance**: The average squared difference from the mean, indicating variability.  
   - **Standard Deviation**: The square root of variance, showing average deviation in the same units as the data.  
     - Example: A lower standard deviation means data points are closer to the mean.

---

### 4. **What is a box plot, and what can it tell you about the distribution of data?**  
   - **Box Plot**: A visual representation of data distribution based on five-number summary (minimum, Q1, median, Q3, maximum).  
   - **Insights**:  
     - Identifies spread, center, and symmetry.  
     - Detects outliers and skewness.

### 5. **Discuss the role of random sampling in making inferences about populations.**  
   - **Random Sampling**: Ensures that every individual in the population has an equal chance of being selected.  
   - **Role**:  
     - Reduces bias.  
     - Allows generalizations about the population.  
     - Ensures the sample is representative.

### 6. **Explain the concept of skewness and its types. How does skewness affect the interpretation of data?**  
   - **Skewness**: A measure of asymmetry in data distribution.  
   - **Types**:  
     - **Positive Skew**: Tail on the right; mean > median.  
     - **Negative Skew**: Tail on the left; mean < median.  
   - **Effect**: Skewness impacts the choice of measures like mean (sensitive to skewness) vs. median (resistant).

### 7. **What is the interquartile range (IQR), and how is it used to detect outliers?**  
   - **IQR**: Difference between Q3 (75th percentile) and Q1 (25th percentile).  
     - Formula: \( \text{IQR} = Q3 - Q1 \).  
   - **Outliers**:  
     - Data points < \( Q1 - 1.5 \times \text{IQR} \) or > \( Q3 + 1.5 \times \text{IQR} \).

### 8. **Discuss the conditions under which the binomial distribution is used.**  
   - **Conditions**:  
     1. Fixed number of trials (\( n \)).  
     2. Two outcomes: success or failure.  
     3. Constant probability of success (\( p \)).  
     4. Independent trials.  
   - Example: Flipping a coin 10 times to count heads.

### 9. **Explain the properties of the normal distribution and the empirical rule (68–95–99.7 rule).**  
   - **Properties**:  
     - Symmetric and bell-shaped.  
     - Mean = Median = Mode.  
     - Defined by mean (\( \mu \)) and standard deviation (\( \sigma \)).  
   - **Empirical Rule**:  
     - 68% of data within \( \mu \pm 1\sigma \).  
     - 95% within \( \mu \pm 2\sigma \).  
     - 99.7% within \( \mu \pm 3\sigma \).

### 10. **Provide a real-life example of a Poisson process and calculate the probability for a specific event.**  
   - **Example**: Number of customer arrivals at a store per hour.  
   - **Probability Formula**:  
     \( P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!} \),  
     where \( \lambda \) = average rate, \( k \) = specific number of events.  
   - If \( \lambda = 5 \) arrivals/hour, \( P(X = 3) \):  
     \( P(X = 3) = \frac{e^{-5} \cdot 5^3}{3!} = 0.1404 \).

### 11. **Explain what a random variable is and differentiate between discrete and continuous random variables.**  
   - **Random Variable**: A variable that takes on outcomes of a random process.  
   - **Discrete**: Countable values (e.g., number of heads in 10 coin flips).  
   - **Continuous**: Infinite values within a range (e.g., weight, height).

### 12. **Provide an example dataset, calculate both covariance and correlation, and interpret the results.**  
   - **Example Dataset**:  
     \( X = [1, 2, 3, 4, 5] \), \( Y = [2, 4, 6, 8, 10] \).  
   - **Covariance**:  
     \( \text{Cov}(X, Y) = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{n-1} \).  
     Result: \( \text{Cov}(X, Y) = 5.0 \).  
   - **Correlation**:  
     \( \text{Corr}(X, Y) = \frac{\text{Cov}(X, Y)}{\sigma_X \cdot \sigma_Y} \).  
     Result: \( \text{Corr}(X, Y) = 1.0 \).  
   - **Interpretation**: Strong positive linear relationship.
