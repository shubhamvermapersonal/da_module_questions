### Theory Questions

#### 1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.
**Answer:**
- **Qualitative Data:** Data that describes qualities or characteristics. Examples: Colors of cars (red, blue), types of fruits (apple, orange).
- **Quantitative Data:** Data that can be measured numerically. Examples: Heights of students, temperatures.
- **Nominal Scale:** Categorical data without a specific order. Example: Gender (male, female).
- **Ordinal Scale:** Categorical data with a meaningful order. Example: Movie ratings (poor, average, good).
- **Interval Scale:** Numeric data with equal intervals but no true zero. Example: Temperature in Celsius.
- **Ratio Scale:** Numeric data with a true zero. Example: Weight of objects.

#### 2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.
**Answer:**
- **Mean:** The average of a dataset. Use when data is symmetrically distributed. Example: Average salary of employees.
- **Median:** The middle value when data is ordered. Use for skewed data or outliers. Example: Median house prices.
- **Mode:** The most frequent value. Use for categorical data. Example: Most common shoe size in a store.

---

#### 3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?
**Answer:**
Dispersion describes how spread out data is. Variance measures the average squared deviation from the mean, while standard deviation (the square root of variance) shows the spread in the same units as the data. Larger values indicate more spread.

#### 4. What is a box plot, and what can it tell you about the distribution of data?
**Answer:**
A box plot is a graphical representation of data that shows the median, quartiles, and potential outliers. It helps identify the data’s spread, skewness, and presence of outliers.

#### 5. Discuss the role of random sampling in making inferences about populations.
**Answer:**
Random sampling ensures that every individual in the population has an equal chance of being selected, minimizing bias and making it possible to generalize results from the sample to the population.

#### 6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
**Answer:**
Skewness measures the asymmetry of a distribution.
- **Positive Skew:** Tail is on the right; mean > median.
- **Negative Skew:** Tail is on the left; mean < median.
Skewness affects the choice of central tendency measures and interpretation of data trends.

#### 7. What is the interquartile range (IQR), and how is it used to detect outliers?
**Answer:**
The IQR is the range between the first quartile (Q1) and the third quartile (Q3). Outliers are values below Q1 – 1.5*IQR or above Q3 + 1.5*IQR.

#### 8. Discuss the conditions under which the binomial distribution is used.
**Answer:**
The binomial distribution is used when:
- There are a fixed number of trials.
- Each trial has two outcomes (success/failure).
- The probability of success is constant.
- Trials are independent.
Example: Flipping a coin 10 times.

#### 9. Explain the properties of the normal distribution and the empirical rule (68–95–99.7 rule).
**Answer:**
The normal distribution is symmetric, bell-shaped, and characterized by its mean and standard deviation.
- **Empirical Rule:**
  - 68% of data falls within 1 SD of the mean.
  - 95% falls within 2 SDs.
  - 99.7% falls within 3 SDs.

#### 11. Explain what a random variable is and differentiate between discrete and continuous random variables.
**Answer:**
A random variable assigns a numerical value to each outcome of a random event.
- **Discrete:** Takes specific values. Example: Number of heads in coin tosses.
- **Continuous:** Takes any value within a range. Example: Heights of students.

#### 12. Provide an example dataset, calculate both covariance and correlation, and interpret the results.
**Answer:**
- **Dataset:** X: [1, 2, 3], Y: [2, 4, 6].
- **Covariance:** Measures how two variables change together.
  Cov(X, Y) = [(1-2)(2-4) + (2-2)(4-4) + (3-2)(6-4)] / 3 = 2.
- **Correlation:** Standardized measure of covariance.
  Correlation = Cov(X, Y) / (σXσY) = 1.
- Interpretation: Positive correlation indicates a strong linear relationship between X and Y.
