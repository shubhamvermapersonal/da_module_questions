# **Theory Questions**

### **1. What is hypothesis testing in statistics?**  
Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves formulating two competing hypotheses:  
- The **null hypothesis (H₀)**, which assumes no effect or no difference.  
- The **alternative hypothesis (H₁)**, which contradicts H₀ and suggests an effect or difference.  

A hypothesis test assesses whether the observed sample data provides enough evidence to reject the null hypothesis. The process includes:  
1. **Defining hypotheses (H₀ and H₁)**  
2. **Selecting a significance level (α), usually 0.05 or 0.01**  
3. **Choosing the appropriate statistical test (Z-test, T-test, etc.)**  
4. **Computing the test statistic**  
5. **Determining the P-value**  
6. **Comparing the P-value with α and making a conclusion**  

If the P-value is smaller than α, we reject H₀ in favor of H₁. Otherwise, we fail to reject H₀.  

---

### **2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**  
| Feature | Null Hypothesis (H₀) | Alternative Hypothesis (H₁) |  
|---------|----------------------|----------------------------|  
| Meaning | Assumes no effect or no difference | Assumes an effect or difference exists |  
| Example | "A new drug has no effect on blood pressure." | "A new drug lowers blood pressure." |  
| Goal | To test if data provides enough evidence to reject it | To determine if there is enough support for a claim |  

The null hypothesis represents the status quo, while the alternative hypothesis challenges it. Statistical tests aim to provide sufficient evidence to reject H₀ in favor of H₁.  

---

### **3. What is the significance level in hypothesis testing, and why is it important?**  
The **significance level (α)** is the probability of rejecting the null hypothesis when it is actually true (Type 1 error). Common values are:  
- **0.05 (5%)**: Moderate evidence required to reject H₀.  
- **0.01 (1%)**: Stronger evidence required to reject H₀.  

The significance level determines the critical region of a test. A lower α reduces the chance of making a Type 1 error but increases the risk of a Type 2 error.  

---

### **4. What does a P-value represent in hypothesis testing?**  
The **P-value** is the probability of obtaining a test result at least as extreme as the observed one, assuming that H₀ is true.  
- A **small P-value (≤ α)** indicates strong evidence against H₀, leading to its rejection.  
- A **large P-value (> α)** suggests insufficient evidence to reject H₀.  

Example:  
If a P-value is **0.03** and α is **0.05**, we reject H₀ because the probability of observing such an extreme result by chance is low.  

---

### **5. How do you interpret the P-value in hypothesis testing?**  
| P-value | Interpretation | Decision (if α = 0.05) |  
|---------|--------------|----------------|  
| **P ≤ 0.01** | Strong evidence against H₀ | Reject H₀ |  
| **0.01 < P ≤ 0.05** | Moderate evidence against H₀ | Reject H₀ |  
| **0.05 < P ≤ 0.10** | Weak evidence against H₀ | Consider rejecting H₀ |  
| **P > 0.10** | No significant evidence | Fail to reject H₀ |  

A P-value does not measure the strength of an effect, only the likelihood that the observed data would occur if H₀ were true.  

---

### **6. What are Type 1 and Type 2 errors in hypothesis testing?**  
| Error Type | Description | Consequence |  
|------------|-------------|--------------|  
| **Type 1 Error (False Positive)** | Rejecting a true null hypothesis | Incorrectly concluding there is an effect |  
| **Type 2 Error (False Negative)** | Failing to reject a false null hypothesis | Missing a real effect |  

Example:  
- **Type 1 Error**: A test wrongly detects a disease in a healthy person.  
- **Type 2 Error**: A test fails to detect a disease in a sick person.  

Reducing Type 1 errors (lowering α) increases Type 2 errors, and vice versa.  

---

### **7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**  
| Test Type | Hypothesis Structure | Application |  
|-----------|----------------------|------------|  
| **One-Tailed Test** | H₁: Mean is **greater** or **less** than a value | Used when testing for an increase/decrease in a specific direction |  
| **Two-Tailed Test** | H₁: Mean is **different** from a value | Used when testing for any significant difference |  

Example:  
- **One-tailed**: "A new fertilizer increases crop yield." (Testing for increase only)  
- **Two-tailed**: "A new fertilizer affects crop yield." (Testing for any change)

### **8. What is the Z-test, and when is it used in hypothesis testing?**  
A **Z-test** is a statistical test used to determine whether two population means are different when the population variance is known and the sample size is large (n ≥ 30).  

#### **When to use a Z-test?**  
- The population standard deviation is known.  
- The sample size is large (Central Limit Theorem applies).  
- The data follows a normal distribution.  

**Example:** Testing if the average height of students differs from 170 cm using a large sample.  

Formula for Z-score:  
\[
Z = \frac{{\bar{X} - \mu}}{{\sigma / \sqrt{n}}}
\]  
where:  
- **X̄** = Sample mean  
- **μ** = Population mean  
- **σ** = Population standard deviation  
- **n** = Sample size  

---

### **9. How do you calculate the Z-score, and what does it represent in hypothesis testing?**  
A **Z-score** represents how many standard deviations a data point is from the mean.  

Formula:  
\[
Z = \frac{{X - \mu}}{{\sigma}}
\]  

#### **Interpretation**  
| Z-Score | Interpretation |  
|---------|---------------|  
| **Z = 0** | Data point is at the mean |  
| **Z = 1** | 1 standard deviation above the mean |  
| **Z = -1** | 1 standard deviation below the mean |  
| **Z = 2 or higher** | Unusual data point |  

In hypothesis testing, a **high Z-score** indicates that the data point is significantly different from the mean, possibly leading to rejecting H₀.  

---

### **10. What is the T-distribution, and when should it be used instead of the normal distribution?**  
The **T-distribution** is a probability distribution used when the population standard deviation is unknown, and the sample size is small (n < 30).  

#### **Comparison: T-distribution vs. Normal Distribution**  
| Feature | Normal Distribution | T-Distribution |  
|---------|----------------------|---------------|  
| Sample Size | Large (n ≥ 30) | Small (n < 30) |  
| Population Std Dev (σ) | Known | Unknown |  
| Shape | Fixed bell curve | Wider tails (more variability) |  

**Example:** If we want to test the average IQ of students using a sample of 20, we use a **T-test** because the standard deviation is unknown.  

Formula for T-score:  
\[
T = \frac{{\bar{X} - \mu}}{{s / \sqrt{n}}}
\]  
where **s** is the sample standard deviation.  

As sample size increases, the **T-distribution** approaches the **Normal distribution**.

### **11. What is the difference between a Z-test and a T-test?**  
| Feature | Z-Test | T-Test |  
|---------|--------|--------|

| Population Standard Deviation (σ) | Known | Unknown |  
| Distribution Used | Normal Distribution | T-Distribution |  
| Purpose | Compares means when σ is known | Compares means when σ is unknown |  

A Z-test is used when we have a large sample and know the population standard deviation, whereas a T-test is used for smaller samples with an unknown standard deviation.

### **12. What is the T-test, and how is it used in hypothesis testing?**  
A **T-test** is a statistical test used to compare the means of two groups when the population standard deviation is unknown.  

#### **Types of T-tests:**  
- **One-sample T-test**: Compares the sample mean to a known population mean.  
- **Two-sample T-test (Independent T-test)**: Compares means of two independent groups.  
- **Paired T-test**: Compares means from the same group before and after a treatment.  

Formula:  
\[
T = \frac{{\bar{X} - \mu}}{{s / \sqrt{n}}}
\]  
where:  
- **X̄** = Sample mean  
- **μ** = Population mean  
- **s** = Sample standard deviation  
- **n** = Sample size  

A large T-value suggests a significant difference between the means.  

---

### **13. What is the relationship between Z-test and T-test in hypothesis testing?**  
Both **Z-test and T-test** are used for hypothesis testing, but their applicability differs:  
- **Z-test** is preferred when σ is known and the sample size is large.  
- **T-test** is used when σ is unknown and the sample size is small.  

As **sample size increases (n → ∞), the T-distribution approaches the Normal (Z) distribution.**

### **14. What is a confidence interval, and how is it used to interpret statistical results?**  
A **confidence interval (CI)** is a range of values that estimates an unknown population parameter with a given level of confidence.  

Formula:  
\[
CI = \bar{X} \pm Z \times \frac{\sigma}{\sqrt{n}}
\]  
where:  
- **X̄** = Sample mean  
- **Z** = Z-score for confidence level  
- **σ** = Standard deviation  
- **n** = Sample size  

For a 95% CI, we are **95% confident that the true mean falls within the interval.**  

---

### **15. What is the margin of error, and how does it affect the confidence interval?**  
The **margin of error (ME)** measures the amount of uncertainty in an estimate.  

Formula:  
\[
ME = Z \times \frac{\sigma}{\sqrt{n}}
\]  

- **A smaller ME** means more precise estimates (requires a larger sample size).  
- **A larger ME** results in a wider confidence interval (less precision).  

For example, if ME is ±2, and X̄ = 50, the CI is **(48, 52).**

Formula:

P(A|B) = \frac{P(B|A) P(A)}{P(B)}
\]  

Applications:  
- **Medical testing** (updating disease probability based on test results).  
- **Spam filtering** (classifying emails as spam or not).  
- **Machine learning** (probabilistic modeling).  

Bayes' Theorem helps in revising prior probabilities with new information.

### **17. What is the Chi-square distribution, and when is it used?**  
The **Chi-square (χ²) distribution** is used for hypothesis testing involving categorical data.  

Formula for Chi-square statistic:  
\[
\chi^2 = \sum \frac{(O - E)^2}{E}
\]  
where **O** = observed frequency, **E** = expected frequency.  

Uses:  
- **Goodness-of-fit test** (checking if data fits a distribution).  
- **Independence test** (checking if two categorical variables are related).  

---

### **18. What is the Chi-square goodness-of-fit test, and how is it applied?**  
The **Chi-square goodness-of-fit test** determines whether sample data fits a theoretical distribution.  

Steps:  
1. Define hypotheses:  
   - **H₀**: The data follows the expected distribution.  
   - **H₁**: The data does not follow the expected distribution.  
2. Compute **expected frequencies**.  
3. Calculate the **Chi-square statistic**.  
4. Compare with the **critical value** (from the Chi-square table).  

Example: Checking if dice rolls follow a uniform distribution.

### **19. What is the F-distribution, and when is it used in hypothesis testing?**  
The **F-distribution** is used to compare variances in two datasets and is right-skewed.  

Applications:  
- **ANOVA (Analysis of Variance)** to compare multiple means.  
- **F-test** to compare variances.  

Formula:  
\[
F = \frac{S_1^2}{S_2^2}
\]  
where **S₁²** and **S₂²** are sample variances.  

If **F > critical value**, variances are significantly different.  

---

### **20. What is an ANOVA test, and what are its assumptions?**  
**ANOVA (Analysis of Variance)** tests if there are significant differences among **three or more** group means.  

#### **Assumptions:**  
1. **Normality**: Data follows a normal distribution.  
2. **Independence**: Samples are independent.  
3. **Equal Variance**: Homogeneity of variances across groups.  

Formula for **F-statistic**:  
\[
F = \frac{\text{Between-group variance}}{\text{Within-group variance}}
\]  

If **F is large**, at least one group mean is significantly different.  

---

### **21. What are the different types of ANOVA tests?**  
| Type | Description | Used When |  
|------|------------|-----------|  
| **One-Way ANOVA** | Compares means of multiple groups | One independent variable |  
| **Two-Way ANOVA** | Examines the effect of two factors | Two independent variables |  
| **Repeated Measures ANOVA** | Compares the same group over time | Measurements taken at multiple time points |  

Example:  
- **One-Way ANOVA**: Comparing test scores across different teaching methods.  
- **Two-Way ANOVA**: Studying the effect of **diet and exercise** on weight loss.  

---

### **22. What is the F-test, and how does it relate to hypothesis testing?**  
An **F-test** compares the variances of two or more groups to determine if they are significantly different.  

Formula:  
\[
F = \frac{S_1^2}{S_2^2}
\]  

| Test | Purpose |  
|------|---------|  
| **F-test for variances** | Compares two sample variances |  
| **ANOVA (F-test for means)** | Compares means of three or more groups |  

If **F > critical value**, reject H₀ (the variances or means are significantly different).

**1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results**

```python
import numpy as np
import scipy.stats as stats

def z_test(sample_data, population_mean, population_std, sample_size):
    sample_mean = np.mean(sample_data)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test
    return z_score, p_value

# Example usage
sample_data = np.random.normal(100, 15, 50)  # Simulated sample data
population_mean = 105
population_std = 15
sample_size = len(sample_data)

z_score, p_value = z_test(sample_data, population_mean, population_std, sample_size)

print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

**5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def type1_type2_errors(sample_size, population_mean, true_mean, alpha=0.05):
    # Type 1 error (False Positive): Rejecting a true null hypothesis
    z_critical = stats.norm.ppf(1 - alpha / 2)  # Two-tailed test
    rejection_region = [-z_critical, z_critical]

    # Simulating two different distributions
    null_dist = stats.norm(population_mean, 1).rvs(sample_size)
    alternative_dist = stats.norm(true_mean, 1).rvs(sample_size)

    # Type 2 error (False Negative): Failing to reject a false null hypothesis
    power = 1 - stats.norm.cdf(z_critical - (true_mean - population_mean) / np.sqrt(sample_size))

    # Plotting the distributions
    x = np.linspace(population_mean - 4, population_mean + 4, 100)
    plt.plot(x, stats.norm.pdf(x, population_mean, 1), label="Null Distribution")
    plt.plot(x, stats.norm.pdf(x, true_mean, 1), label="Alternative Distribution", linestyle='--')

    plt.fill_between(x, stats.norm.pdf(x, population_mean, 1), where=(x < rejection_region[0]) | (x > rejection_region[1]), color='red', alpha=0.3, label="Type 1 Error Region")

    plt.legend()
    plt.title("Type 1 and Type 2 Errors")
    plt.show()

    return power, alpha

# Example usage
sample_size = 30
population_mean = 100
true_mean = 105
power, alpha = type1_type2_errors(sample_size, population_mean, true_mean)
print(f"Power of the test: {power}")
```

**6. Write a Python program to perform an independent T-test and interpret the results.**

```python
import numpy as np
import scipy.stats as stats

def independent_t_test(sample1, sample2):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    return t_stat, p_value

# Example usage
sample1 = np.random.normal(100, 15, 50)  # Group 1
sample2 = np.random.normal(105, 15, 50)  # Group 2

t_stat, p_value = independent_t_test(sample1, sample2)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

**9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.**

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data)
    sample_size = len(sample_data)
    z_critical = stats.norm.ppf((1 + confidence_level) / 2)

    margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
sample_data = np.random.normal(100, 15, 50)

lower, upper = confidence_interval(sample_data)

print(f"Confidence Interval: ({lower}, {upper})")
```

**11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.**

```python
def bayes_theorem(prior, likelihood, evidence):
    # Bayes' Theorem: P(H|E) = (P(E|H) * P(H)) / P(E)
    posterior = (likelihood * prior) / evidence
    return posterior

# Example usage
prior = 0.6  # P(H) (prior probability of hypothesis)
likelihood = 0.8  # P(E|H) (probability of evidence given hypothesis)
evidence = 0.7  # P(E) (probability of evidence)

posterior = bayes_theorem(prior, likelihood, evidence)
print(f"Posterior Probability: {posterior}")
```

**13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.**

```python
import numpy as np

def expected_frequencies(observed_data):
    row_totals = np.sum(observed_data, axis=1)
    col_totals = np.sum(observed_data, axis=0)
    total = np.sum(observed_data)

    expected = np.outer(row_totals, col_totals) / total
    return expected

# Example usage
observed_data = np.array([[30, 10], [20, 40]])
expected = expected_frequencies(observed_data)
print(f"Expected Frequencies: \n{expected}")
```

**16. Implement an F-test using Python to compare the variances of two random samples.**

```python
import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    f_stat = var1 / var2 if var1 > var2 else var2 / var1
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1
    p_value = 1 - stats.f.cdf(f_stat, df1, df2)
    return f_stat, p_value

# Example usage
sample1 = np.random.normal(100, 15, 50)
sample2 = np.random.normal(100, 20, 50)

f_stat, p_value = f_test(sample1, sample2)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")
```

**17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.**

```python
import numpy as np
import scipy.stats as stats

def anova_test(groups):
    f_stat, p_value = stats.f_oneway(*groups)
    return f_stat, p_value

# Example usage
group1 = np.random.normal(100, 15, 50)
group2 = np.random.normal(110, 15, 50)
group3 = np.random.normal(120, 15, 50)

f_stat, p_value = anova_test([group1, group2, group3])

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis: At least one group mean is different.")
else:
    print("Fail to reject the null hypothesis: All group means are equal.")
```

**19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA**

```python
import numpy as np
import scipy.stats as stats

def check_anova_assumptions(groups):
    # Check normality assumption using Shapiro-Wilk test
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        if p_value < 0.05:
            print(f"Group {i+1} does not follow a normal distribution.")
        else:
            print(f"Group {i+1} follows a normal distribution.")

    # Check equality of variances using Levene's test
    stat, p_value = stats.levene(*groups)
    if p_value < 0.05:
        print("Variances are not equal.")
    else:
        print("Variances are equal.")

    # Check independence (this is typically done by design, no formal test)
    print("Assumed that groups are independent by design.")

# Example usage
group1 = np.random.normal(100, 15, 50)
group2 = np.random.normal(110, 15, 50)
group3 = np.random.normal(120, 15, 50)

check_anova_assumptions([group1, group2, group3])
```

**20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.**

**24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.**

```python
import numpy as np
import scipy.stats as stats

def chi_square_variance_test(sample_data, population_variance):
    sample_size = len(sample_data)
    sample_variance = np.var(sample_data, ddof=1)

    chi2_stat = (sample_size - 1) * sample_variance / population_variance
    p_value = 1 - stats.chi2.cdf(chi2_stat, sample_size - 1)

    return chi2_stat, p_value

# Example usage
sample_data = np.random.normal(100, 15, 50)
population_variance = 225

chi2_stat, p_value = chi_square_variance_test(sample_data, population_variance)

print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")
```

**25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.**

```python
import numpy as np
import scipy.stats as stats

def z_test_proportions(successes1, n1, successes2, n2):
    p1 = successes1 / n1
    p2 = successes2 / n2
    p_pool = (successes1 + successes2) / (n1 + n2)

    z_score = (p1 - p2) / np.sqrt(p_pool * (1 - p_pool) * (1 / n1 + 1 / n2))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

    return z_score, p_value

# Example usage
successes1 = 45
n1 = 100
successes2 = 55
n2 = 100

z_score, p_value = z_test_proportions(successes1, n1, successes2, n2)

print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")
```

**26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.**
