# **Theory Questions**

### **1. What is a random variable in probability theory?**  
A **random variable (RV)** is a numerical value assigned to the outcomes of a random experiment. It allows us to quantify uncertainty in probability theory.  
- **Example:** Tossing a coin can have two outcomes: heads (H) or tails (T). If we define a random variable \(X\) as:  
  - \(X = 1\) if heads appears  
  - \(X = 0\) if tails appears  
  Here, \(X\) represents the outcome numerically.  
- **Types:** Discrete and Continuous (explained in the next section).  

---

### **2. What are the types of random variables?**  
Random variables are classified into:  

| Type              | Definition | Example |
|------------------|-----------|---------|
| **Discrete RV**   | Takes countable values | Number of defective items in a batch (0, 1, 2, …). |
| **Continuous RV** | Takes an infinite number of values in an interval | Temperature in a city (e.g., 27.5°C, 28.2°C, etc.). |

---

### **3. What is the difference between discrete and continuous distributions?**  
A **probability distribution** describes how probabilities are assigned to different values of a random variable.  

| Feature | Discrete Distribution | Continuous Distribution |
|---------|----------------------|----------------------|
| Definition | Probability is assigned to individual values | Probability is assigned to a range of values |
| Example | Rolling a die (1, 2, 3, 4, 5, 6) | Heights of people (can take any real number in a range) |
| Probability Calculation | \(P(X = x)\) directly gives the probability | Uses a probability density function (PDF) |

---

### **4. What are probability distribution functions (PDFs)?**  
A **Probability Distribution Function (PDF)** describes the probability of different values of a random variable.  
- **For discrete variables:** It is called a **probability mass function (PMF)** and gives probabilities for each discrete value.  
- **For continuous variables:** It is a function where probabilities are found by integrating over an interval.  

Example:  
- **Discrete Case:** If rolling a fair die, the probability of getting any one number (1–6) is **\(1/6\)**.  
- **Continuous Case:** For heights, we cannot assign probabilities to exact values but can find probabilities over ranges (e.g., probability of a person being between 170-175 cm).  

---

### **5. How do cumulative distribution functions (CDFs) differ from PDFs?**  
A **Cumulative Distribution Function (CDF)** gives the probability that a random variable takes a value **less than or equal** to a given number.  

| Feature | PDF (PMF for discrete) | CDF |
|---------|------------------|------------------|
| Definition | Probability of a specific value | Probability of a value less than or equal to a number |
| Representation | Function \( P(X = x) \) | Function \( P(X \leq x) \) |
| Example | \( P(X = 3) = 0.2 \) | \( P(X \leq 3) = P(X=1) + P(X=2) + P(X=3) \) |

For a continuous variable, the **CDF is the integral of the PDF**:  
\[
F(x) = \int_{-\infty}^{x} f(t) dt
\]  

---

### **6. What is a discrete uniform distribution?**  
A **discrete uniform distribution** assigns equal probabilities to **finite, discrete** outcomes.  

**Formula:**  
\[
P(X = x) = \frac{1}{n}, \quad x = x_1, x_2, ..., x_n
\]  
where \( n \) is the total number of possible outcomes.  

Example: Rolling a fair six-sided die:  
- \( P(X = 1) = P(X = 2) = P(X = 3) = P(X = 4) = P(X = 5) = P(X = 6) = \frac{1}{6} \).  

This distribution is useful in situations where all outcomes are equally likely, such as lotteries or card shuffling.  

---

### **7. What are the key properties of a Bernoulli distribution?**  
A **Bernoulli distribution** represents a random experiment with **only two possible outcomes**, often referred to as "success" and "failure."  

**Properties:**  
- Takes values **0 or 1**.  
- Probability of success \( P(X=1) = p \), failure \( P(X=0) = 1 - p \).  
- **Mean** \( E[X] = p \).  
- **Variance** \( Var(X) = p(1 - p) \).  

**Example:** Tossing a coin where  
- \( X = 1 \) for heads (success), \( X = 0 \) for tails (failure).  
- If the probability of heads is **\( p = 0.5 \)**, then:  
  - \( P(X=1) = 0.5 \),  
  - \( P(X=0) = 1 - 0.5 = 0.5 \).  

---

### **8. What is the binomial distribution, and how is it used in probability?**  
The **binomial distribution** models the number of successes in **\( n \)** independent **Bernoulli** trials.  

**Formula:**  
\[
P(X = k) = \binom{n}{k} p^k (1 - p)^{n-k}
\]  
where:  
- \( n \) = total trials  
- \( k \) = successes  
- \( p \) = probability of success  
- \( \binom{n}{k} \) = \( \frac{n!}{k!(n-k)!} \)  

**Example:**  
- Flipping a coin 10 times, finding the probability of exactly 5 heads.  

It is widely used in **quality control, A/B testing, and medical trials**.  

---

### **9. What is the Poisson distribution and where is it applied?**  
A **Poisson distribution** models the number of events in a fixed interval **when they occur randomly and independently**.  

**Formula:**  
\[
P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}
\]  
where \( \lambda \) is the **average rate** of occurrences.  

**Examples:**  
- Number of customer arrivals at a shop per hour.  
- Number of emails received per minute.  
- Number of errors in a page of a book.  

It is widely used in fields like **network traffic analysis, call centers, and epidemiology**.  

---

### **10. What is a continuous uniform distribution?**  
A **continuous uniform distribution** describes a situation where any value in an interval is equally likely.  

**Formula (PDF):**  
\[
f(x) = \frac{1}{b-a}, \quad a \leq x \leq b
\]  
where \( a \) and \( b \) define the range.  

**Properties:**  
- **Mean** \( E[X] = \frac{a + b}{2} \).  
- **Variance** \( Var(X) = \frac{(b - a)^2}{12} \).  

**Example:** The probability of a bus arriving between **10:00 AM and 10:30 AM** assuming it arrives **randomly** is uniform over the 30-minute period.  

This distribution is widely used in **simulation modeling and scheduling problems**.

### **11. What are the characteristics of a normal distribution?**  
A **normal distribution** (or Gaussian distribution) is a continuous probability distribution that is symmetric around the mean.  

**Key Characteristics:**  
- **Bell-shaped curve**: The distribution is symmetric around its mean.  
- **Mean, Median, and Mode are equal**: The peak of the curve occurs at the mean.  
- **Defined by two parameters**: Mean (\(\mu\)) and standard deviation (\(\sigma\)).  
- **68-95-99.7 Rule (Empirical Rule)**:  
  - 68% of data falls within \( 1\sigma \) of the mean.  
  - 95% within \( 2\sigma \).  
  - 99.7% within \( 3\sigma \).  
- **Asymptotic**: The tails of the distribution extend indefinitely.  

**Example:** Heights of people, IQ scores, and test scores follow a normal distribution.  

---

### **12. What is the standard normal distribution, and why is it important?**  
The **standard normal distribution** is a special case of the normal distribution where:  
- Mean \( \mu = 0 \).  
- Standard deviation \( \sigma = 1 \).  

**Why is it important?**  
- Any normal distribution can be converted to a standard normal distribution using a **Z-score transformation**:  
  \[
  Z = \frac{X - \mu}{\sigma}
  \]  
- The **Z-table** helps calculate probabilities directly.  
- It simplifies hypothesis testing and confidence interval calculations.  

**Example:** In SAT scores, a raw score of 1300 can be converted to a **Z-score** to determine how well a student performed compared to others.

### **13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**  
The **Central Limit Theorem (CLT)** states that the sampling distribution of the sample mean approaches a normal distribution **regardless of the population’s original distribution**, as long as the sample size is sufficiently large (\( n \geq 30 \)).  

**Importance:**  
- **Enables statistical inference**: Even if data is not normally distributed, we can make predictions using normal distribution formulas.  
- **Forms the basis of hypothesis testing and confidence intervals**.  
- **Allows approximation of probabilities for sums and averages of random variables**.  

**Example:** If we sample 50 people from a population of all students, their **average height distribution** will be approximately normal, even if the original population’s height distribution is skewed.  

---

### **14. How does the Central Limit Theorem relate to the normal distribution?**  
- The **CLT justifies the use of normal distribution** in many real-world scenarios.  
- The sample mean \(\bar{X}\) follows a normal distribution even if the population itself is not normal.  
- As the sample size increases, the standard error \( \sigma_{\bar{X}} = \frac{\sigma}{\sqrt{n}} \) decreases, making the sample mean more precise.  

**Example:** If we repeatedly sample groups of 50 students and calculate their average test scores, the distribution of these sample means will be normal.

### **15. What is the application of Z statistics in hypothesis testing?**  
**Z-statistics** are used in hypothesis testing when the population standard deviation (\(\sigma\)) is known and the sample size is large (\( n \geq 30 \)).  

**Applications:**  
- **Testing population means**: Compare sample mean to a population mean.  
- **Proportion tests**: Used in comparing categorical data proportions.  
- **Quality control**: Ensuring manufacturing processes meet specifications.  

**Example:** A factory claims that light bulbs last **1,000 hours**. A sample of **50 bulbs** is tested, and the average lifespan is **980 hours**. A **Z-test** determines whether the deviation is statistically significant.

### **16. How do you calculate a Z-score, and what does it represent?**  
A **Z-score** (standard score) tells how many standard deviations a value is from the mean.  

**Formula:**  
\[
Z = \frac{X - \mu}{\sigma}
\]  
where:  
- \(X\) = observed value  
- \( \mu \) = mean  
- \( \sigma \) = standard deviation  

**Interpretation:**  
- \( Z > 0 \) → Value is above the mean.  
- \( Z < 0 \) → Value is below the mean.  
- \( Z = 1.96 \) → 95% of data lies below this value.  

**Example:** If the average test score is 70 with \( \sigma = 10 \), and a student scores 85:  
\[
Z = \frac{85 - 70}{10} = 1.5
\]  
This means the student scored **1.5 standard deviations above the mean**.  

---

### **17. What are point estimates and interval estimates in statistics?**  
- **Point estimate**: A single value used to estimate a population parameter (e.g., sample mean \(\bar{X}\) for population mean \(\mu\)).  
- **Interval estimate**: A range of values with a confidence level (e.g., **confidence interval (CI)**).  

**Example:**  
- **Point estimate:** The average height of a sample is **5.6 feet**, estimating the population mean.  
- **Interval estimate:** "The population mean height is **between 5.5 and 5.7 feet** with **95% confidence**."  

---

### **18. What is the significance of confidence intervals in statistical analysis?**  
A **confidence interval (CI)** provides a range within which a population parameter likely falls.  

**Formula for CI:**  
\[
\bar{X} \pm Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}
\]  
where \(Z_{\alpha/2}\) is the critical Z-score for the confidence level.  

**Example:** If a sample mean exam score is **75** and the 95% CI is **\(75 \pm 3\)**, then the true population mean is likely between **72 and 78**.  

---

### **19. What is the relationship between a Z-score and a confidence interval?**  
- A **Z-score** determines how far a value is from the mean.  
- Confidence intervals are constructed using **Z-scores** for a specific confidence level (e.g., **1.96** for 95%).  

**Example:** In a **95% confidence interval**, a Z-score of **\(\pm 1.96\)** means the estimate covers 95% of the population data.  

---

### **20. How are Z-scores used to compare different distributions?**  
Z-scores standardize different datasets, allowing for comparisons across different distributions.  

**Example:**  
- Student A scores **85** in math (\(\mu = 70, \sigma = 10\)), Z = 1.5.  
- Student B scores **90** in physics (\(\mu = 75, \sigma = 15\)), Z = 1.0.  
- Even though Student B's raw score is higher, Student A performed better relative to their peers.  

---

### **21. What are the assumptions for applying the Central Limit Theorem?**  
1. **Random Sampling**: Samples must be randomly selected.  
2. **Independence**: Observations must be independent.  
3. **Sample Size**: \( n \geq 30 \) for non-normal populations.  
4. **Finite Variance**: Population variance must be finite.  

---

### **22. What is the concept of expected value in a probability distribution?**  
The **expected value (mean)** represents the long-term average outcome of a random variable.  

**Formula:**  
\[
E[X] = \sum X P(X)
\]  
for discrete variables, or  
\[
E[X] = \int x f(x) dx
\]  
for continuous variables.  

---

### **23. How does a probability distribution relate to the expected outcome of a random variable?**  
- The **probability distribution** describes all possible values and their probabilities.  
- The **expected value** provides the long-run average when repeating the experiment.  

**Example:** In rolling a fair die,  
\[
E[X] = (1+2+3+4+5+6)/6 = 3.5
\]  
Even though **3.5 never occurs**, it is the expected long-run average.

**6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete

**9. Write a Python function to calculate Z-scores from a dataset and plot them.**

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    return (data - mean) / std_dev

# Generate dataset
data = np.random.normal(50, 10, 1000)  # Mean=50, Std Dev=10
z_scores = calculate_z_scores(data)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='orange', edgecolor='black')
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.title("Histogram of Z-scores")
plt.show()
```

**15. Implement hypothesis testing using Z-statistics for a sample dataset.**

```python
import scipy.stats as stats
import numpy as np

# Example data
sample = np.random.normal(100, 15, 30)  # Sample from normal distribution
pop_mean = 105  # Population mean (null hypothesis)

# Perform Z-test
z_score = (np.mean(sample) - pop_mean) / (np.std(sample, ddof=1) / np.sqrt(len(sample)))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

print(f"Z-score: {z_score}, P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

**16. Create a confidence interval for a dataset using Python and interpret the result.**
