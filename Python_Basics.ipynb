### 1. Explain the key features of Python that make it a popular choice for programming.

### 2. Describe the role of predefined keywords in Python and provide examples of how they are used in a program.

Predefined keywords are reserved words in Python with special meanings. These keywords cannot be used as variable names. They ensure consistency and readability in the code. Some examples include:

- **if, else**: Used for control flow.
  ```python
  if x > 0:
      print("Positive")
  ```
- **def**: Used to define functions.
  ```python
  def greet():
      return "Hello!"
  ```
- **class**: Used to define classes.
  ```python
  class MyClass:
      pass
  ```
- **True, False**: Boolean literals representing truth values.
  
These keywords help structure Python code and allow the language to perform specific tasks in a standardized way.

### 3. Compare and contrast mutable and immutable objects in Python with examples.

- **Mutable**: Objects whose values can change after creation.
  - **Example**: Lists are mutable.
    ```python
    my_list = [1, 2, 3]
    my_list[0] = 10  # Modifies the list
    ```
  
- **Immutable**: Objects whose values cannot change after creation.
  - **Example**: Strings and tuples are immutable.
    ```python
    my_string = "hello"
    my_string[0] = "H"  # Error: Strings are immutable
    ```

### 4. Discuss the different types of operators in Python and provide examples of how they are used.

Python operators are used to manipulate values. They are classified as:

- **Arithmetic Operators**: Used to perform mathematical operations.
  ```python
  5 + 3  # Output: 8
  ```
- **Relational Operators**: Used to compare values.
  ```python
  5 > 3  # Output: True
  ```
- **Logical Operators**: Used to combine conditional statements.
  ```python
  True and False  # Output: False
  ```
- **Bitwise Operators**: Used to manipulate bits of data.
  ```python
  5 & 3  # Output: 1
  ```
- **Assignment Operators**: Used to assign values to variables.
  ```python
  x = 10  # Assign 10 to x
  x += 2  # x = x + 2, so x becomes 12
  ```

### 5. Explain the concept of type casting in Python with examples.

Type casting refers to converting one data type to another. It can be explicit or implicit.

- **Explicit Casting**: Done manually by the programmer.
  ```python
  x = int("10")  # Converts string "10" to integer 10
  ```

- **Implicit Casting**: Handled automatically by Python.
  ```python
  x = 10 + 5.5  # Integer is automatically converted to float (result: 15.5)
  ```

### 6. How do conditional statements work in Python? Illustrate with examples.

Conditional statements control the flow of execution based on certain conditions.

#### Example:
```python
x = 10
if x > 5:
    print("Greater")
elif x == 5:
    print("Equal")
else:
    print("Lesser")
```
This program prints `"Greater"` because the condition `x > 5` is `True`.

### 7. Describe the different types of loops in Python and their use cases with examples.

Python provides two main types of loops for iteration:

- **For Loop**: Iterates over a sequence (e.g., a list, tuple, or range).
  ```python
  for i in range(5):
      print(i)
  ```
  Use case: Best for known or fixed iterations.

- **While Loop**: Continues running as long as the specified condition is `True`.
  ```python
  x = 0
  while x < 5:
      print(x)
      x += 1
  ```
  Use case: Ideal for scenarios where the number of iterations is not known in advance, but a condition must be met.
