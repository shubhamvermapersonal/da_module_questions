### 1. Discuss string slicing and provide examples.

String slicing extracts portions of a string using index positions. Python uses the format `string[start:end:step]`.

#### Example:
```python
s = "Python"
print(s[1:4])  # Output: "yth"
print(s[:3])   # Output: "Pyt"
print(s[::2])  # Output: "Pto"
```

String slicing simplifies tasks like reversing (`s[::-1]`) or extracting substrings.

### 2. Explain the key features of lists in Python.

Lists in Python are versatile and ordered collections of elements. Key features include:

- **Mutable**: Elements can be changed.
- **Dynamic**: Size can vary during runtime.
- **Heterogeneous**: Supports mixed data types (e.g., `[1, "hello", 3.5]`).
- **Built-in Methods**: Functions like `append()`, `pop()`, and `sort()`.

### 3. Describe how to access, modify, and delete elements in a list with examples.

- **Access**: Use indices (`my_list[0]`).
- **Modify**: Assign new values (`my_list[0] = "new_value"`).
- **Delete**: Use `del`, `remove()`, or `pop()`.

#### Example:
```python
my_list = [1, 2, 3]
my_list[1] = 10  # Modify
del my_list[0]   # Delete
```

### 4. Compare and contrast tuples and lists with examples.

- **Tuples**: Immutable, defined with parentheses `(1, 2, 3)`. Faster due to immutability.
- **Lists**: Mutable, defined with square brackets `[1, 2, 3]`. Suitable for dynamic changes.

#### Example:
```python
t = (1, 2, 3)  # Tuple: Cannot modify
l = [1, 2, 3]  # List: Can modify
l[0] = 10
```

### 5. Describe the key features of sets and provide examples of their use.

Sets are unordered, unique collections. Key features include:

- **No Duplicates**: Automatically removes duplicates (e.g., `{1, 2, 2}` becomes `{1, 2}`).
- **Unordered**: No indexing.
- **Operations**: Support union, intersection, and difference.

#### Example:
```python
s = {1, 2, 3}
s.add(4)  # Adds 4 to the set
```

### 8. Discuss the importance of dictionary keys being immutable and provide examples.

Dictionary keys must be immutable (e.g., strings, tuples) to ensure hash-based access. Mutable keys, like lists, cause errors as they can change and disrupt lookups.

#### Example:
```python
valid_dict = {("a", "b"): 1}  # Tuple key
invalid_dict = {[1, 2]: "value"}  # Error: Lists cannot be dictionary keys
```
